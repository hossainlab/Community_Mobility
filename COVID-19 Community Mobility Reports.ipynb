```python
import pandas as pd
import plotly.express as px 
# download the csv file from here: https://www.google.com/covid19/mobility/index.html?hl=en
df = pd.read_csv('Global_Mobility_Report.csv')
```

### Cleaning steps

```python
# We only focus on the trend in the U.S
df = df.loc[df['country_region']=="Bangladesh"]
# Drop the columns that we will not use for the analysis
df = df.drop(columns=['country_region_code','country_region','iso_3166_2_code', 'census_fips_code', 
                      "sub_region_1", "sub_region_2", 'metro_area', 'place_id'])
# Rename the rest of the columns for easier accessin
df.rename(columns={
                        'retail_and_recreation_percent_change_from_baseline': 'retail',
                        'grocery_and_pharmacy_percent_change_from_baseline': 'grocery',
                        'parks_percent_change_from_baseline': 'parks',
                        'transit_stations_percent_change_from_baseline': 'transit',
                        'workplaces_percent_change_from_baseline': 'workplaces',
                        'residential_percent_change_from_baseline': 'residential'}, inplace=True)
```

```python
df.head() 
```

```python
df.isnull().sum() 
```

```python
df.columns = df.columns.str.title() 
```

```python
df.columns
```

```python
df.to_csv("global_mobility_clean_data.csv", index=False)
```

```python
fig = px.line(df, x="date", y=['retail'], 
                labels= {"value": "Changes", 
                        'retail': 'Mobility Trend for Retail'})
fig.update_layout(
    title = "Community Mobility Reports of Bangladesh (Data: https://www.google.com/covid19/mobility/)", 
    xaxis = dict(title = "Date")
)
fig.show()
```

```python
fig = px.line(df, x="Date", y=['Retail', 'Grocery', 'Parks', 'Transit', 'Workplaces',
       'Residential'], 
                labels= {"value": "Changes", 'variable': 'Place Categories'})
fig.update_layout(
    xaxis = dict(title = "Date"), 
    font_size = 16
)
fig.update_xaxes(title_font_family="Arial")
fig.show()
```
